```python
department_count=HR['Department'].value_counts()
print("department_count:",department_count)
sns.barplot(x=department_count.index,y=department_count.values,palette='Set2')
plt.title("DEPARTMENTS COUNTS")
plt.xticks(rotation=45)
plt.show()
```

```python
job_count=HR['Job_Title'].value_counts()
print("Jobs Count",job_count)
plt.figure(figsize=(12,6))
sns.barplot(x=job_count.index,y=job_count.values,color='green')
plt.title("jobs_count")
plt.xticks(rotation=90)
plt.show()
```

```python
Avg_salary=HR.groupby(by='Department')['Salary_INR'].mean()/1000
Avg_salary = Avg_salary.sort_values(ascending=False)
print("Avg_salary:",Avg_salary)
plt.figure(figsize=(15,6))
sns.barplot(x=Avg_salary.index,y=Avg_salary.values,color='r')
plt.title("AVG SALARY")
plt.xticks(rotation=90)
plt.show()

```

```python
Avg_salary=HR.groupby(by='Job_Title')['Salary_INR'].mean()/1000
Avg_salary = Avg_salary.sort_values(ascending=False)
print("Avg_salary:",Avg_salary)
plt.figure(figsize=(15,6))
sns.barplot(x=Avg_salary.index,y=Avg_salary.values,color='c')
plt.title("AVG SALARY")
plt.xticks(rotation=90)
plt.show()
```

```python
terminated_counts = HR[HR['Status'] == 'Terminated'].groupby('Department').size()
print(terminated_counts)
sns.barplot(x=terminated_counts.index,y=terminated_counts.values)
plt.title("TERMINTED COUNT PERDEPARTMENT")
plt.xticks(rotation=45)
plt.show()
```

```python
resigned_counts = HR[HR['Status'] == 'Resigned'].groupby('Department').size()
print("Resigned count per Department:")
print(resigned_counts)
sns.barplot(x=resigned_counts.index,y=resigned_counts.values)
plt.title("Resigned count per Department")
plt.xticks(rotation=45)
plt.show()
```

```python
HR['Hire_Date'] = pd.to_datetime(HR['Hire_Date'], errors='coerce')
count_emp=HR.groupby(HR['Hire_Date'].dt.year)['Employee_ID'].count()
print("YEAR WISE HIRED EMPLOYEE:")
print(count_emp)
plt.figure(figsize=(8, 5))
sns.lineplot(x=count_emp.index, y=count_emp.values,color='red')  
plt.xlabel('Year')
plt.ylabel('Number of Hires')
plt.title('Year-wise Hired Employees')
plt.grid(True)
plt.show()
```
